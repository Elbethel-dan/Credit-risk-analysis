### Defining Features and Target Variable

```python
# Define features and target
X = df.drop(columns=[TARGET_COL, ID_COL])
y = df[TARGET_COL]
```

### Train All Models

```python
print("ðŸš€ Training models...")

metrics = train_model(X, y)

print("\nðŸŽ‰ Training complete. Evaluation metrics:")

for model_name, metric_vals in metrics.items():
    print(f"\n{model_name}:")
    for k, v in metric_vals.items():
        print(f"  {k}: {v:.4f}")
```

- Based on the ROC-AUC scores, Logistic Regression achieved the highest discriminative performance (0.9882), slightly outperforming XGBoost (0.9868) and Random Forest (0.9412). Decision Tree performed poorly in distinguishing classes, with an ROC-AUC near 0.5, indicating it is almost random despite high accuracy due to class imbalance. Overall, Logistic Regression is the best model for this dataset in terms of ROC-AUC.

```python
# Prepare features for prediction (exclude target & ID)
X_pred = df.drop(columns=[TARGET_COL, ID_COL])

# Apply the same preprocessing
X_pred_processed = clean_feature_names(X_pred)
```

```python
best_model_name = max(metrics, key=lambda m: metrics[m]["roc_auc"])
print(f"\nðŸ† Best model based on ROC-AUC: {best_model_name}")
```
